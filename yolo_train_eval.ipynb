```python
from ultralytics import YOLO
model = YOLO('/content/drive/MyDrive/model.pt')
```

```python
results = model.val(data='/content/drive/MyDrive/Blood Cell detection.v1i.yolov12/data.yaml')
```

```python
metrics_dict = results.results_dict  

print(metrics_dict)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

metrics_dict = {
    'metrics/precision(B)': 0.8080822056139819,
    'metrics/recall(B)': 0.892497025795111,
    'metrics/mAP50(B)': 0.8969916883590701,
    'metrics/mAP50-95(B)': 0.6045956315680894,
    'fitness': 0.6338352372471875
}

metrics_df = pd.DataFrame(list(metrics_dict.items()), columns=['Metric', 'Value'])

print(metrics_df)

sns.barplot(x='Metric', y='Value', data=metrics_df[metrics_df['Metric'].str.contains('mAP50')])
plt.title('mAP50 values')
plt.show()

sns.barplot(x='Metric', y='Value', data=metrics_df)
plt.title('Evaluation Metrics')
plt.xticks(rotation=45)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow  


img_path = '/content/drive/MyDrive/Blood Cell detection.v1i.yolov12/test/images/BloodImage_00007_jpg.rf.18eaea0a5abcc0657c54038656aa3857.jpg'
img = cv2.imread(img_path)  

results = model(img)  

print(type(results)) 
print(results)

if isinstance(results, list):
    result = results[0]  
    boxes = result.boxes  

    for bbox in boxes:
        x1, y1, x2, y2 = bbox.xyxy[0]  
        conf = bbox.conf[0]  
        cls = int(bbox.cls[0]) 

        cv2.rectangle(img, (int(x1), int(y1)),
                      (int(x2), int(y2)), (0, 255, 0), 2) 
        cv2.putText(img, f'Class {cls} Conf: {conf:.2f}', (int(x1), int(y1) - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('Detection Results')
plt.axis('off')  
plt.show()

cv2_imshow(img)
```

```python
test_metrics_dict = {
    'metrics/precision(B)': 0.8080822056139819,
    'metrics/recall(B)': 0.892497025795111,
    'metrics/mAP50(B)': 0.8969916883590701,
    'metrics/mAP50-95(B)': 0.6045956315680894,
    'fitness': 0.6338352372471875
}

val_results = model.val(data='/content/drive/MyDrive/Blood Cell detection.v1i.yolov12/valdata.yaml')
val_metrics_dict = val_results.results_dict  

print("Validation Metrics:")
print(val_metrics_dict)
print("\nTest Metrics:")
print(test_metrics_dict)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

val_df = pd.DataFrame(list(val_metrics_dict.items()), columns=['Metric', 'Validation'])
test_df = pd.DataFrame(list(test_metrics_dict.items()), columns=['Metric', 'Test'])

comparison_df = pd.merge(val_df, test_df, on='Metric', how='outer')

print(comparison_df)

```

```python
sns.barplot(x='Metric', y='Validation', data=comparison_df, color='blue', label='Validation')
sns.barplot(x='Metric', y='Test', data=comparison_df, color='orange', label='Test')

plt.title('Comparison of Validation and Test Metrics')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend()

plt.show()


```
